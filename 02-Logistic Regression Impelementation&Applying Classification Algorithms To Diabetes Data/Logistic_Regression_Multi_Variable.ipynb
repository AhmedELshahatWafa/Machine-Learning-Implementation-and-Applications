```python
def encode_columns(df_data):
    
    # replacing zeros with medians 
    
#     skin_median =df_data['SkinThickness'].median() 
#     print(skin_median)
#     df_data.loc[df_data['SkinThickness'] == 0, 'SkinThickness'] = skin_median
    
    blood_median =df_data['BloodPressure'].median() 
    #print(blood_median)
    df_data.loc[df_data['BloodPressure'] == 0, 'BloodPressure'] = blood_median
    
#     insulin_median =df_data['Insulin'].median() 
#     print(insulin_median)
#     df_data.loc[df_data['Insulin'] == 0, 'Insulin'] = insulin_median
    
    bmi_median =df_data['BMI'].median() 
    #print(bmi_median)
    df_data.loc[df_data['BMI'] == 0, 'BMI'] = bmi_median
    
    df_data = df_data.drop(columns = ["SkinThickness","Insulin"])
    
    
    return df_data
```

```python
df_data = encode_columns(df_data)
```

```python
def sigmoid(x):
    return 1 / (1+math.e**((-x)))
```

```python
def GradientDecentMulti_Logistic(input_matrix,outputVector,alpha=0.002,n_iter= 10000):
    m = len(input_matrix)
    lst_thetas = np.zeros(len(input_matrix[0])) 
    learning_Rate = alpha
    loss_vals = []
    
    for i in range(n_iter):
        
        #gettin the predicted data vector 
        predict_data_vector = sigmoid( np.dot(input_matrix,lst_thetas) )
        #GRADIENT VECTOR
        grad_vector = np.dot(input_matrix.T,(predict_data_vector - outputVector))/m
        #updating thetas
        lst_thetas = lst_thetas -  learning_Rate* grad_vector

    print(lst_thetas)
    return lst_thetas 


def predict(lst_thetas, input_matrix):
    return sigmoid(input_matrix @ ls_thetas)
```

```python
y =np.array(df_data[["Outcome"]]).reshape(1,-1)[0]
```

```python
ls_thetas =  GradientDecentMulti_Logistic(arr,y)
```

```python
predicted = np.round(predict(ls_thetas, arr))
```
